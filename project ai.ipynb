```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

we are using handwritten data set from keras
and by using this line "keras.datasets.mnist.load_data()" we load train and test data test into these blow variables

```python
(X_train, y_train) , (X_test, y_test) =  keras.datasets.mnist.load_data()
```

```python
# we 60000 digits images in this variable
len(X_train)
```

```python
#similarly
len(X_test)
```

we  have 28/28 pixel image in sample

```python
X_train[0].shape
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
plt.matshow(X_train[0])
```

```python
y_train[0]
```

```python
#shape(items,rows.columns)
X_train.shape
```

now we are going to convert 2D array into single dimensional array, which have 784 elements

```python
# we are using reshape function from pandas
X_train_flattened = X_train.reshape(len(X_train),28*28)
print("shape of 'X_train_flattened' is",X_train_flattened.shape)

X_test_flattened = X_test.reshape(len(X_test),28*28)
print("shape of 'X_test_flattened' is",X_test_flattened.shape)
```

```python
X_train_flattened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,),activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[7])
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[7]
```

```python
np.argmax(y_predicted[7])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,),activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
model.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flattened,y_test)
```
